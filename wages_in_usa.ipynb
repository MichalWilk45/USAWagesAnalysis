```python
sns.lineplot(x=wages['year'], y=wages['median'], label='median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['average'], label='average')
plt.legend()
plt.show()
```

```python
sns.lineplot(x=wages['year'], y=wages['men_median'], label='men_median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['women_median'], label='women_median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['men_average'], label='men_average')
sns.lineplot(x=wages['year'], y=wages['women_average'], label='women_average')
plt.legend()
plt.show()
```

Median pay for men is only one that did not increase substantially over the years. Although men average increased

Difference between Median and Average Pay between Women and Men

```python
sns.lineplot(x=wages['year'], y=(wages['men_median'] - wages['women_median'])/wages['men_median']*100, label='median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=(wages['men_average'] - wages['women_average'])/wages['men_average']*100, label='average')
plt.ylabel('Difference [%]')
plt.xlabel('Year')
plt.legend()
plt.show()
```

Difference between paygap has decreased over las 50 years but still oon average women earn less then men 

**Ethicity comparision**

```python
sns.lineplot(x=wages['year'], y=wages['white_median'], label='white median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['black_median'], label='black median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['hispanic_median'], label='hispanic median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['white_average'], label='white average')
sns.lineplot(x=wages['year'], y=wages['black_average'], label='black average')
sns.lineplot(x=wages['year'], y=wages['hispanic_average'], label='hispanic average')

plt.ylabel('Wage')
plt.xlabel('Year')
plt.legend()
plt.show()
```

```python
sns.lineplot(x=wages['year'], y=wages['white_men_median'], label='white men median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['black_men_median'], label='black men median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['hispanic_men_median'], label='hispanic men median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['white_men_average'], label='white men average')
sns.lineplot(x=wages['year'], y=wages['black_men_average'], label='black men average')
sns.lineplot(x=wages['year'], y=wages['hispanic_men_average'], label='hispanic men average')

plt.ylabel('Wage')
plt.xlabel('Year')
plt.legend()
plt.show()
```

```python
sns.lineplot(x=wages['year'], y=wages['white_women_median'], label='white women median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['black_women_median'], label='black women median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['hispanic_women_median'], label='hispanic women median', linestyle="dashed")
sns.lineplot(x=wages['year'], y=wages['white_women_average'], label='white women average')
sns.lineplot(x=wages['year'], y=wages['black_women_average'], label='black women average')
sns.lineplot(x=wages['year'], y=wages['hispanic_women_average'], label='hispanic women average')

plt.ylabel('Wage')
plt.xlabel('Year')
plt.legend()
plt.show()
```

**Change between median and average salary over the years**

```python
sns.lineplot(x=wages['year'], y=(wages['average']-wages['median'])/wages['average']*100, label='whole population')

plt.legend()
plt.xlabel('Year')
plt.ylabel('Difference Average vs Median[%]')
plt.show()
```

This shows that wages distribution is becoming increasingly right skewed. People who earn more has haster rate of increasing their wages.

```python
sns.lineplot(x=wages['year'], y=(wages['men_average']-wages['men_median'])/wages['men_average']*100, label='men')
sns.lineplot(x=wages['year'], y=(wages['women_average']-wages['women_median'])/wages['women_average']*100, label='women')

plt.legend()
plt.xlabel('Year')
plt.ylabel('Difference Average vs Median[%]')
plt.show()
```

Difference between MEdian and Average raises fater for men than women

```python
sns.lineplot(x=wages['year'], y=(wages['white_average']-wages['white_median'])/wages['white_average']*100, label='white')
sns.lineplot(x=wages['year'], y=(wages['black_average']-wages['black_median'])/wages['black_average']*100, label='black')
sns.lineplot(x=wages['year'], y=(wages['hispanic_average']-wages['hispanic_median'])/wages['hispanic_average']*100, label='hispanic')

plt.legend()
plt.xlabel('Year')
plt.ylabel('Difference Average vs Median[%]')
plt.show()
```

Difference between ethnicity grups is really small and all follow similar trend
